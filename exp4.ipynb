```python
set6 = {1,2,3,4}
set7 = {3,4,5,6}

print("union:", set6 | set7)
#or
print("union:", set6.union(set7))

print("intersection:", set6 & set7)
#or
print("intersection:", set6.intersection(set7))

print("difference:", set6 - set7)
print("difference:", set7 - set6)
print("dsimilar:", set6 ^ set7)
```

```python
#que 2

student1 = {"Python", "Data Structures", "Algorithms", "Networking"}
student2 = {"Python", "Operating Systems", "Algorithms", "Data Structures"}
student3 = {"Python", "Machine Learning", "Algorithms", "Cyber Security"}

common_electives = student1 & student2 & student3

print("Student 1 Electives:", student1)
print("\nStudent 2 Electives:", student2)
print("\nStudent 3 Electives:", student3)
print("\n\nCommon subjects chosen by all students:", common_electives)
```
